```python
times = pd.DatetimeIndex(df1.index)
grouped = df.groupby([times.year]).mean()
```

```python
grouped.head()
```

```python
plt.figure(figsize= (15,5))
plt.plot(grouped['AAPL'])
plt.title("Average Land Temperature 1750-2015")
plt.xlabel("year")
plt.ylabel("Adj Closing price")
plt.show()
```

```python
x = grouped.index.values.reshape(-1, 1)
y = grouped['AAPL'].values
```

```python
reg = LinReg()
reg.fit(x,y, 1)
y_pred = reg.predict(x)
print("Accuracy: " + str(reg.score(x, y)))
```

```python
plt.figure(figsize = (15, 5))
plt.title('Linear Regression')
plt.scatter(x=x, y=y_pred)
plt.scatter(x=x, y=y, c="b")
```
